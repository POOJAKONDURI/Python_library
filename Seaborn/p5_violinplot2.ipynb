```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from the provided URL
url = 'https://github.com/mwaskom/seaborn-data/raw/master/iris.csv'
data = pd.read_csv(url)

# Create a violin plot of 'species' against 'sepal_length'
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=data)

# Add titles and labels
plt.title('Violin Plot of Species vs Sepal Length')
plt.xlabel('Species')
plt.ylabel('Sepal Length')

# Show the plot
plt.show()
```
