```python
import matplotlib.pyplot as plt
import numpy as np

# Group A (Heights and Weights)
heights_A = np.random.normal(165, 10, 50)
weights_A = np.random.normal(65, 8, 50)

# Group B 
heights_B = np.random.normal(175, 10, 50)
weights_B = np.random.normal(75, 8, 50)

# Group C 
heights_C = np.random.normal(185, 10, 50)
weights_C = np.random.normal(85, 8, 50)

# Create the scatter plot for Group A, B, and C
plt.scatter(heights_A, weights_A, color='blue', label='Group A', alpha=0.6)
plt.scatter(heights_B, weights_B, color='green', label='Group B', alpha=0.6)
plt.scatter(heights_C, weights_C, color='red', label='Group C', alpha=0.6)

plt.xlabel('Heights (cm)')
plt.ylabel('Weights (kg)')
plt.title('Scatter Plot Comparing Weights and Heights of Three Groups')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
