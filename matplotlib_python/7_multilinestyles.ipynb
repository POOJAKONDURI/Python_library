```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 1, 2, 3, 4]

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("multiline different styles")

plt.plot(x,y1,label = "quadratuc",linestyle = "--",color='r')
plt.plot(x,y2,label = "linear",linestyle = "-",color = 'b')

plt.legend()
plt.show()
```
