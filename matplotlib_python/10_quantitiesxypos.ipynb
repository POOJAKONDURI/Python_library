```python
import matplotlib.pyplot as plt

# Data (X, Y positions)
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

a = [11,12,3,14,5]
b = [21,10,22,14,4]

# Plotting the data points
plt.plot(x, y, marker='o', color='b', linestyle='--')
plt.plot(a ,b, marker='x', color='r', linestyle='--')

# Adding labels and title
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Plotting Quantities with X and Y Positions')

# Display the plot
plt.show()
```
