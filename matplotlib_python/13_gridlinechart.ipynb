```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'Date': ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16'],
        'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]}
df = pd.DataFrame(data)
plt.plot(df['Date'], df['Close'], marker='o', color='b', label='Close Price')

plt.grid(True, which='both', linestyle='-', linewidth=0.5, color='blue')
plt.title('Alphabet Inc. Closing Price (Oct 3 - Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
