```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.figure(figsize=(8, 6))
plt.subplot(3, 1,2)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([9, 10,15, 200])
plt.figure(figsize=(8, 6))
plt.subplot(3, 1, 2)
plt.plot(x,y)


plt.show()
```
