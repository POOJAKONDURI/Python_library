```python
import matplotlib.pyplot as plt

# Example: Line Chart
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Chart Example')
plt.show()

#bar chart
import matplotlib.pyplot as plt

plt.bar([0.25,2.25,3.25,5.25,7.25],[300,400,200,600,700],
label="Car",color='b')
plt.bar([0.75,1.75,2.75,3.75,4.75],[50,30,20,50,60],
label="bike", color='g')
plt.legend()
plt.xlabel('km')
plt.ylabel('milege')
plt.title('Details')
plt.show()
# scater plot
import matplotlib.pyplot as pyplot

x1 = [1, 2.5,3,4.5,5,6.5,7]
y1 = [1,2, 3, 2, 1, 3, 4]
x2=[8, 8.5, 9, 9.5, 10, 10.5, 11]
y2=[3,3.5, 3.7, 4,4.5, 5, 5.2]
pyplot.scatter(x1, y1, label = 'car', color='c')
pyplot.scatter(x2,y2,label='bike',color='g')
pyplot.title('vehicle')
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.legend()
pyplot.show()

#pie chart
import matplotlib.pyplot as pyplot
slice = [12, 25, 50, 36, 19]
activities = ['NLP','Neural Network', 'Data analytics', 'Quantum Computing', 'Machine Learning']
cols = ['r','b','c','g', 'orange']
pyplot.pie(slice,
labels =activities,
colors = cols)
pyplot.title('Training Subjects')
pyplot.show()

#histogram
import matplotlib.pyplot as pyplot
x = [22,55,62,45,21,22,34,42,42,4,2,8]
y = [1,10,20,30,40,50]
pyplot.hist(x, y, rwidth=0.6)
pyplot.xlabel('age groups')
pyplot.ylabel('Number of people')
pyplot.title('Histogram')
pyplot.show()
```
