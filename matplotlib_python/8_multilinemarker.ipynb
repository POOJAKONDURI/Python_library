```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 1, 2, 3, 4]
plt.plot(x, y1, label="Quadratic", marker='o', linestyle='--', color='r')  # Circle marker
plt.plot(x, y2, label="Linear", marker='x', linestyle='-', color='b')      # Cross marker


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines with Markers')


plt.legend()

plt.show()
```
