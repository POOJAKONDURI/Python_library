```python
import matplotlib.pyplot as plt
x = []
y = []
with open('test.txt') as f:
    for line in f:
        data = line.split(' ')
        x.append(int(data[0]))
        y.append(int(data[1]))
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot from textfile")
plt.show()
                 


```
