```python
import matplotlib.pyplot as plt
import numpy as np

# Data
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
sd_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

# Create indices for the x-axis
indices = np.arange(len(mean_velocity))

# Create the bar plot with error bars
plt.bar(indices, mean_velocity, yerr=sd_velocity, capsize=5, color='blue', alpha=0.7, label='Mean Velocity')

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Mean Velocity')
plt.title('Bar Plot with Error Bars')

# Set x-ticks
plt.xticks(indices, ['1', '2', '3', '4'])

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
