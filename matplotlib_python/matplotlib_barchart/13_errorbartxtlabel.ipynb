```python
import matplotlib.pyplot as plt
import numpy as np

# Data
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
sd_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

# Create indices for the x-axis
indices = np.arange(len(mean_velocity))

# Create the bar plot with error bars
bars = plt.bar(indices, mean_velocity, yerr=sd_velocity, capsize=5, color='blue', alpha=0.7, label='Mean Velocity')

# Add a text label above each bar displaying the integer value of the mean velocity
for i, bar in enumerate(bars):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.02, f'{int(yval*1000)}', ha='center', va='bottom')

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Mean Velocity')
plt.title('Bar Plot with Error Bars and Text Labels')

plt.xticks(indices, ['1', '2', '3', '4'])
plt.legend()
plt.show()
```
