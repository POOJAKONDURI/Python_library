```python

import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {'a':[2, 4, 6, 8, 10],
        'b': [4, 8, 5, 7, 6],
        'c': [2, 3, 4, 2, 6],
        'd': [4, 7, 4, 7, 8],
        'e': [2, 6, 4, 8, 6]
        }

df = pd.DataFrame(data) 


# Create a bar plot
df.plot.bar()

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Bar Plot from DataFrame')

# Show the plot
plt.show()
```
