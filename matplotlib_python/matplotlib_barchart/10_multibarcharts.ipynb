```python
import matplotlib.pyplot as plt
import numpy as np
group = ['g1','g2','g3','g4','g5']
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
X_axis = np.arange(len(group)) 
  
plt.bar(X_axis - 0.2, men, 0.4, label = 'Girls') 
plt.bar(X_axis + 0.2, women, 0.4, label = 'Boys') 
  
plt.xticks(X_axis, group) 
plt.xlabel("Groups") 
plt.ylabel("scores by group") 
plt.title("gender in each group") 
plt.legend() 
plt.show() 
```
