```python
import matplotlib.pyplot as plt

# Sample data
programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7,6.7]
 

# Specify bar width and alignment
bar_width = 0.35
bar_positions = [i + bar_width / 2 for i in range(len(programming_languages))]

# Create the bar chart
plt.bar(bar_positions, popularity, width=bar_width, color='blue')

# Set labels and title
plt.xticks(range(len(programming_languages)), programming_languages)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')

# Show the plot
plt.show()
```
