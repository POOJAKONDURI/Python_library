```python
'''

    @Author:pooja
    @Date:20-09-2024
    @last modified by:pooja
    @last modified time:20-09-2024
    @title:1. Write a program to draw a scatter plot for random 1000 x and y coordinates

'''
import plotly.express as px
import pandas as pd
import numpy as np
x  = np.random.rand(1000)
y = np.random.rand(1000)
df = pd.DataFrame({'X':x,'Y':y})

fig = px.scatter(df,x='X',y='Y', title='Scatter Plot of Random 1000 Coordinates')
fig.show()
```
