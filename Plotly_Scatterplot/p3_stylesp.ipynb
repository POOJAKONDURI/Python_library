```python
'''

    @Author:pooja
    @Date:20-09-2024
    @last modified by:pooja
    @last modified time:20-09-2024
    @title: program to draw a scatter plot for random 500 x and y coordinates and style it

'''

import numpy as np
import plotly.express as px
import pandas as pd

# Generate random coordinates
x = np.random.rand(500)
y = np.random.rand(500)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Create a styled scatter plot
fig = px.scatter(data, x='X', y='Y', 
                 title='Styled Scatter Plot of Random 500 Coordinates',
                 color_discrete_sequence=['purple'], 
                 size_max=20)

# Update marker properties
fig.update_traces(marker=dict(line=dict(width=2, color='hotpink')))

# Show the plot
fig.show()
```
