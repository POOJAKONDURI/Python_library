```python
'''

    @Author:pooja
    @Date:20-09-2024
    @last modified by:pooja
    @last modified time:20-09-2024
    @title:2. Write a program to draw line and scatter plots for random 100 x and y coordinates

'''
import numpy as np
import plotly.express as px
import pandas as pd

# Generate random coordinates
x = np.random.rand(100)
y = np.random.rand(100)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Create a scatter plot and line plot using Plotly Express
fig = px.scatter(data, x='X', y='Y', title='Line and Scatter Plots of Random 100 Coordinates', color_discrete_sequence=['red'])

# Add a line plot by using the same data
fig.add_scatter(x=data['X'], y=data['Y'], mode='lines', name='Line', line=dict(color='blue'))

# Show the plot
fig.show()

```
